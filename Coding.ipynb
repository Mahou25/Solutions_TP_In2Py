from requests.exceptions import HTTPError


for url in ['https://api.github.com', 'https://api.github.com/invalid']:

<h1>urllib.robotparser</h1>

<p>Ce module contient une seule classe, RobotFileParser. Cette classe rpond  la question de savoir si un utilisateur particulier peut ou non rcuprer une URL qui a publi des fichiers robot.txt. Robots.txt est un fichier texte que les webmasters crent pour indiquer aux robots Web comment explorer les pages de leur site Web. Le fichier robot.txt indique au racleur Web quelles parties du serveur ne doivent pas tre accessibles.
Par exemple :</p>

```python
from urllib.robotparser import*
import urllib.robotparser as rb 
  
bot = rb.RobotFileParser() 
  
x = bot.set_url('https://www.geeksforgeeks.org / robot.txt') 
print(x) 
  
y = bot.read() 
print(y) 
  
z = bot.can_fetch('*', 'https://www.geeksforgeeks.org/') 
print(z) 
  
w = bot.can_fetch('*', 'https://www.geeksforgeeks.org / wp-admin/') 
print(w) 
```
